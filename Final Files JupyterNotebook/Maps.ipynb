```python
######################## READING COMBINED TABLE CREATED DURING DATA CLEANING ####################

combinedtable = "Resources/Combined Tables_OneFile/FlightDataCLEANwithLatLong.csv"
combinedtable = pd.read_csv(combinedtable,encoding="utf-8")
combinedtable.info()
```

```python
# CONVERT LAT TO FLOAT BECAUSE IT WAS AN OBJECT 
convert_lat = {'Lat': float}
combinedtable = combinedtable.astype(convert_lat)
combinedtable.info()
```

```python
# GROUP DATA BY CITY TO GET QUOTES COUNT
combinedtablegrouped = combinedtable.groupby("City Name")
combinedtablegrouped.head()
```

```python
# OBTAIN QUOUTES COUNT
quotes_counts = combinedtablegrouped['Quote ID'].count()
quotes_counts = quotes_counts.reset_index() 
#quotes_counts_list = quotes_counts.tolist()
#quotes_counts_list

# Data frame to dictionary. I realize that I don't need to do this
#https://www.geeksforgeeks.org/python-pandas-dataframe-to_dict/
#quotes_counts_dict = quotes_counts.to_dict()

# I just drop the duplicates but keep the first city of each row 
combinedtable_drop_cities = combinedtable.drop_duplicates(subset = "City Name", keep = 'first')

# Merge quotes and drop cities DF considering that they both are the same shape
tableheatmap = pd.merge(combinedtable_drop_cities,quotes_counts, how ='outer', on = 'City Name')

# Rename second Quote ID to to Total Quotes
tableheatmap = tableheatmap.rename(columns ={'Quote ID_y': 'Total Quotes'})
tableheatmap.info()
```

```python
# Store latitude and longitude in locations
locations2 = tableheatmap[["Lat", "Long"]]
locations2.info()

#Store quotes in total quotes
total_quotes = tableheatmap['Total Quotes']
total_quotes.head()
```
